```python
import pandas as pd
df = pd.read_csv('score.csv')
df.head(5)
```

```python
print("Statistika Jam Belajar:")
print(df['Hours'].describe())
print("\nStatistika Nilai:")
print(df['Scores'].describe())
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x = df[['Hours']]
y = df['Scores']

model = LinearRegression()
model.fit(x,y)

beta_0 = model.intercept_
beta_1 = model.coef_[0]

print(f"Persamaan Regresi : Nilai = {beta_0:.2f} + {beta_1:.2f} * Jam Belajar")
#ketika seseorangg tidak belajar sama sekali (0 jam), model memprediksi nilainya sekitar 2.48
#setiap tambahan 1 jam belajar akan menambah nilai sebesar 9.78
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='Hours', y ='Scores', data=df, ci=None, line_kws={"color":"green"})
plt.title('Hubungan Jam Belajar dan Nilai diberbagai kota', fontsize=16)
plt.xlabel('Jam Belajar', fontsize=12)
plt.ylabel('Nilai', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.annotate(f'Nilai = {beta_0:.2f} + {beta_1:.2f} * Jam',
             xy=(0.55,0.9), xycoords='axes fraction', fontsize=12,color='blue')
plt.show()
```
